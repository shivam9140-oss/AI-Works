```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
#importing the data ;)
path="https://drive.google.com/uc?export=download&id=1UwpW8GELggsVNj1IHtAo5rm4js_6b6PZ"
data = pd.read_csv(path)
#spilliting the data, to train and to test later ;)
X_train, X_test, y_train, y_test = train_test_split(data.drop('variety', axis=1), data['variety'], test_size=0.3, random_state=42)

#Training a random forest classifier to classify our data
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)

#Testing -> tested on sample x_test
y_pred = rfc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Model accuracy: {accuracy:.3f}') # 1.000 accuracy (excelent)
```
